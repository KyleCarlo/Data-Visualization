# Visualizing the Data

Libraries needed.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

## Fuel Industry and Other Industry CO2 Emission

```python
world_emission = pd.read_csv('rendered/factors/world_co2.csv')
```

```python
min_year_emission = min(world_emission['Year'])
max_year_emission = max(world_emission['Year'])

# getting the maximum range for y
max_yaxis_emission = round(max(world_emission.Value) + 1000)
bar_chart = px.bar(world_emission, x='Factor', y='Value', color='Factor',
                   animation_frame='Year', range_y=[0, max_yaxis_emission],
                  )
bar_chart.update_layout(
    template=dict(
        updatemenus=[
            dict(
                type='buttons',
                showactive=False,
                buttons=[
                    dict(
                        label='Play',
                        method='animate',
                        args=[None, dict(frame=dict(duration=500), fromcurrent=True,
                                          transition=dict(duration=0),
                                          mode='immediate')]
                    ),
                    dict(
                        label='Pause',
                        method='animate',
                        args=[[None], dict(frame=dict(duration=0), mode='immediate', transition=dict(duration=0))]
                    )
                ],
                direction='left',
                pad=dict(r=10, t=87),
                x=0.1,
                y=0,
                bgcolor='#292929',  # set the background color of the button
                bordercolor='white',  # set the border color of the button
                font=dict(color='white')  # set the font color of the button
            )
        ]
    )
)
bar_chart.show()
# date_selected = 2021
# emission_year = world_emission[world_emission['Year'] == date_selected][['Cement CO2', 'Coal CO2', 'Gas CO2', 'Oil CO2', 'Flaring CO2', 'Other Industry CO2']].transpose().reset_index()
# emission_year.rename(columns = {emission_year.columns[0] : 'Factors', emission_year.columns[1] : 'Values'}, inplace=True)
# emission_year.sort_values('Values', inplace=True, ascending=False)
# Go counterpart
# bar_chart = go.Figure([go.Bar(x=emission_year['Factors'], y=emission_year['Values'])])
# bar_chart.update_layout(yaxis_range=[0, max_yaxis_emission])
# bar_chart.show()
```
